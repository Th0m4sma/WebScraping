```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
from time import sleep

# Configurações do ChromeDriver
chrome_options = Options()
chrome_options.add_argument("--disable-gpu")
chrome_options.add_argument("--no-sandbox")
chrome_options.headless = False

# Inicializa o WebDriver
driver = webdriver.Chrome(options=chrome_options)

try:
    # Carrega a página inicial
    driver.get("https://exame.com/economia/ultimas-noticias/1/")
    print("Página carregada.")

    # Aguarda o carregamento inicial da página
    sleep(5)

    # Remove camadas que possam bloquear o clique
    driver.execute_script("""
        var elementos = document.querySelectorAll('div, iframe');
        for (var i = 0; i < elementos.length; i++) {
            if (elementos[i].style.zIndex > 0 || elementos[i].style.position === 'fixed') {
                elementos[i].parentNode.removeChild(elementos[i]);
            }
        }
    """)
    print("Camadas bloqueadoras removidas.")

    # Verifica e clica no botão "Next"
    botao_principal = WebDriverWait(driver, 15).until(
        EC.element_to_be_clickable((By.CSS_SELECTOR, "a[rel='next']"))
    )

    # Rolagem para o botão (garantir visibilidade)
    driver.execute_script("arguments[0].scrollIntoView();", botao_principal)
    sleep(1)

    # Tenta clicar via JavaScript
    driver.execute_script("arguments[0].click();", botao_principal)
    print("Clique realizado via JavaScript.")

    # Aguarda a nova página carregar após o clique
    sleep(3)

except Exception as e:
    print(f"Erro ao interagir com a página: {e}")
    driver.save_screenshot("erro_botao.png")  # Salva screenshot para depuração

finally:
    # Finaliza o WebDriver
    driver.quit()
    print("Driver finalizado.")
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
from time import sleep
import pandas as pd

# Configurações do ChromeDriver
chrome_options = Options()
chrome_options.add_argument("--headless")  # Ativa o modo headless
chrome_options.add_argument("--disable-gpu")
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--disable-blink-features=AutomationControlled")
chrome_options.add_argument(
    "user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36"
)

# Inicia o driver
driver = webdriver.Chrome(options=chrome_options)
driver.get("https://economia.uol.com.br/noticias/")

i = 1

while True:
    sleep(10)
    try:
        # Aguarda o botão principal estar presente
        botao_principal = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.XPATH, "//button[contains(text(), 'Ver mais')]"))
        )

        driver.execute_script("arguments[0].scrollIntoView();", botao_principal)
        
        # Remova iframes para evitar bloqueios
        driver.execute_script("""
            var iframes = document.querySelectorAll('iframe');
            for (var i = 0; i < iframes.length; i++) {
                iframes[i].parentNode.removeChild(iframes[i]);
            }
        """)
        
        print("Todos os iframes foram removidos.")
        print(f"Clique de número ({i}).")
        i += 1
        
        # Realiza o clique no botão
        driver.execute_script("arguments[0].click();", botao_principal)

    except Exception as e:
        print("Erro ao localizar ou interagir com o botão: (Chegamos no final)")
        break

sleep(8)
lista = []

# Coleta e analisa o conteúdo da página
conteudo = BeautifulSoup(driver.page_source, "html.parser")
particao_div = conteudo.find_all("div", class_="flex-wrap")

print("Quantidade de divs: ", len(particao_div))

for divs in particao_div:
    elementos_a = divs.find_all("a")
    print("Quantidade: ", len(elementos_a))

    for item in elementos_a:
        elementos_txt = item.find("h3", class_="thumb-title title-xsmall title-lg-small")
        elementos_data = item.find("time", class_="thumb-date")

        titulo = elementos_txt.get_text(strip=True) if elementos_txt else "Título não encontrado"
        data = elementos_data.get_text(strip=True) if elementos_data else "Data não disponível"

        print("Título: ", titulo)
        print("Link: ", item.get("href"))
        print("Data: ", data)
        lista.append([titulo, item.get("href"), data])

driver.quit()

# Salva os dados no CSV
data = pd.DataFrame(lista, columns=["Manchete", "Link", "Data"])
data.to_csv("dadosUOL.csv", index=False)
print(data.head())

```
